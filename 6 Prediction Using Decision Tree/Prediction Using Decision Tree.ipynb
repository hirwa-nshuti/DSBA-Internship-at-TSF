Importing the Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Import the model and an additional visualization tool.
from sklearn.tree import DecisionTreeClassifier, plot_tree

import warnings
warnings.filterwarnings(action="ignore")
```

### Loading the dataset

```python
# Loading the iris dataset
iris=pd.read_csv('Iris.csv')
iris.head()
```

Plotting the Pairplot to see how data behaves

```python
import seaborn as sns
sns.pairplot(iris, hue = 'Species',diag_kind="hist")
```

Remove the column containing the target name since it doesn't contain numeric values.

```python
X = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
X[0:5]
```

```python
y=iris['Species']
y.head()
```

```python
# Split the data into training and testing subsets
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                        random_state=1)
print('Shape of X training set {}'.format(X_train.shape),'&',
      ' Size of Y training set {}'.format(y_train.shape))
print('Shape of X training set {}'.format(X_test.shape),'&',
      ' Size of Y training set {}'.format(y_test.shape))
```

### Building the Model

```python
# Model Object
classifier = DecisionTreeClassifier()

# Model Training
classifier = classifier.fit(X_train, y_train)

# Prediction
y_pred = classifier.predict(X_test)
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score

# Model accuracy
eval_model = accuracy_score(y_test, y_pred)

print("accuracy : " , eval_model)
```

## Ploting the decision Tree

#### Function to plot the Tree

```python
def figu(model):
    # Plotting tree
    plt.figure(figsize = (16,10))
    plot_tree(model,
          rounded = True,
          filled = True)
    # Displaying the figure
    plt.show()
```

Plotting the trained Model

```python
#initializing the model object

model_=classifier.fit(X,y)

#plotting the tree
figu(model_)
```

Using max_deph

```python
# model with max_deph=3
model_3 = DecisionTreeClassifier(max_depth = 3).fit(X, y)

# Plotting tree
figu(model_3)
```

Plotting initial dataset with minimum impurity

```python
# model with a minimum impurity decrease of 0.01
model_imp = DecisionTreeClassifier(min_impurity_decrease = 0.01).fit(X, y)

#Plotting tree
figu(model_imp)
```
